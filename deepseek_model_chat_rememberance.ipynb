```python
import sys
import ollama
from PyQt5.QtWidgets import (
    QApplication, QWidget, QVBoxLayout,
    QTextEdit, QLineEdit, QPushButton,
    QHBoxLayout
)

class ChatApp(QWidget):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("DeepSeek Chat")
        self.setGeometry(200, 200, 500, 600)

        self.messages = []          # conversation memory
        self.remember_mode = True   # default mode

        main_layout = QVBoxLayout()
        button_layout = QHBoxLayout()

        # Chat display
        self.chat_display = QTextEdit()
        self.chat_display.setReadOnly(True)

        # Input
        self.input_box = QLineEdit()

        # Buttons
        self.send_button = QPushButton("Send")
        self.remember_button = QPushButton("Remember Chat")
        self.nomatter_button = QPushButton("No Matter")

        # Add buttons to horizontal layout
        button_layout.addWidget(self.remember_button)
        button_layout.addWidget(self.nomatter_button)

        # Add widgets
        main_layout.addWidget(self.chat_display)
        main_layout.addLayout(button_layout)
        main_layout.addWidget(self.input_box)
        main_layout.addWidget(self.send_button)

        self.setLayout(main_layout)

        # Connections
        self.send_button.clicked.connect(self.send_message)
        self.input_box.returnPressed.connect(self.send_message)
        self.remember_button.clicked.connect(self.enable_memory)
        self.nomatter_button.clicked.connect(self.disable_memory)

    def enable_memory(self):
        self.remember_mode = True
        self.chat_display.append("🔵 Memory Mode Enabled\n")

    def disable_memory(self):
        self.remember_mode = False
        self.chat_display.append("🟠 No Memory Mode Enabled\n")

    def send_message(self):
        user_text = self.input_box.text()
        if not user_text:
            return

        self.chat_display.append(f"You: {user_text}")
        self.input_box.clear()

        if self.remember_mode:
            # Add user message to history
            self.messages.append(
                {"role": "user", "content": user_text}
            )

            response = ollama.chat(
                model="deepseek-r1:8b",
                messages=self.messages
            )

            bot_reply = response["message"]["content"]

            # Add bot reply to history
            self.messages.append(
                {"role": "assistant", "content": bot_reply}
            )

        else:
            # No memory mode (fresh every time)
            response = ollama.chat(
                model="deepseek-r1:8b",
                messages=[
                    {"role": "user", "content": user_text}
                ]
            )

            bot_reply = response["message"]["content"]

        self.chat_display.append(f"Bot: {bot_reply}\n")


if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = ChatApp()
    window.show()
    sys.exit(app.exec_())
```
