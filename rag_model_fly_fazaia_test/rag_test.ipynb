# CREATING DATA (ALDEARY CREATED, DO NOT RUN AGAIN)

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
```

```python
# Load document
loader = TextLoader("fly_fazaia_dataset.txt")
documents = loader.load()

```

```python
# Split text into chunks
splitter = CharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

```

```python
docs = splitter.split_documents(documents)

# Create embeddings
embeddings = HuggingFaceEmbeddings()

# Create vector database
db = Chroma.from_documents(
    docs,
    embeddings,
    persist_directory="fly_db"
)
```

```python
db.persist()

print("Database created successfully.")
```

# USING MODEL

```python
from langchain_community.vectorstores import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
import ollama
```

```python
# Load database
embeddings = HuggingFaceEmbeddings()

db = Chroma(
    persist_directory="fly_db",
    embedding_function=embeddings
)
```

```python
def ask(question):

    docs = db.similarity_search(question, k=3)

    context = "\n".join([d.page_content for d in docs])

    prompt = f"""
        Use the following context to answer the question.
        
        Context:
        {context}
        
        Question:
        {question}
        """

    response = ollama.chat(
        model="llama3.1:8b",
        messages=[{"role":"user","content":prompt}]
    )

    return response["message"]["content"]
```

```python
print(ask("What courses does Fly Fazaia offer?"))
```

```python
print(ask("where is this organization located?"))
```

```python
print(ask("Is fly fazaia a good organization and i can enroll myself in it?"))
```
